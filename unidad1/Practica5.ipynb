```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# cargar el archivo CSV en una variable llamada covid
covid = pd.read_csv("covid.zip")
```

```python
# mostramos los primeros 5 registros del archivo
covid.head()
```

```python
covid.info()
```

```python
# mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
covid.EDAD.min()
```

```python
media = covid.EDAD.mean()

plt.pyplot.hist(covid.EDAD, bins = 7)
plt.pyplot.title("histograma edad pruebas COVID")
plt.pyplot.axvline(media, color='y', linestyle='dashed',
                   linewidth=2, label=f'media={media:.2f}')
plt.pyplot.xlabel("edad")
plt.pyplot.ylabel("frecuencia")
```

```python
plt.pyplot.hist(covid.SEXO, bins = 3)
plt.pyplot.title("histograma sexo pruebas COVID")

plt.pyplot.xlabel("sexo")
plt.pyplot.ylabel("frecuencia")
```

```python
plt.pyplot.hist(covid.TIPO_PACIENTE, bins = 7)
plt.pyplot.title("histograma tipo de paciente pruebas COVID")
plt.pyplot.xlabel("tipo de paciente")
plt.pyplot.ylabel("frecuencia")
```

```python
# crear un diagram de caja para visulizar la distribucion
sns.boxplot(x="SEXO", y="EDAD", data=covid)
```
