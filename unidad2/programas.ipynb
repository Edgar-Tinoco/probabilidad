# probabilidad y estadisticas unidad 2

```python
import math
from itertools import combinations, combinations_with_replacement

def combinaciones_sin_repeticion():
    print("\n--- COMBINACIONES SIN REPETICIÓN ---")
    # Pedimos al usuario los valores de n y r
    n = int(input("Ingrese el número total de elementos (n): "))
    r = int(input("Ingrese el número de elementos por grupo (r): "))

    # Fórmula: C(n, r) = n! / [r!(n-r)!]
    resultado = math.comb(n, r)

    print(f"El número de combinaciones C({n}, {r}) es: {resultado}")

def ejemplo_letras_numeros():
    print("\n--- EJEMPLO CON LETRAS Y NÚMEROS ---")
    # Pedimos una cadena de caracteres
    caracteres = input("Ingrese los caracteres (por ejemplo: ABC123): ")
    r = int(input("Ingrese el número de elementos por grupo (r): "))

    # Generamos las combinaciones
    lista = list(combinations(caracteres, r))

    print(f"\nSe pueden formar {len(lista)} combinaciones:")
    for c in lista:
        print("".join(c))

def combinaciones_con_repeticion():
    print("\n--- COMBINACIONES CON REPETICIÓN ---")
    # Pedimos datos al usuario
    elementos = input("Ingrese los elementos (por ejemplo: ABC): ")
    r = int(input("Ingrese el número de elementos por grupo (r): "))

    # Generamos combinaciones con repetición
    lista = list(combinations_with_replacement(elementos, r))

    print(f"\nSe pueden formar {len(lista)} combinaciones con repetición:")
    for c in lista:
        print("".join(c))

# Programa principal
while True:
    print("\n=== MENÚ PRINCIPAL ===")
    print("1. Combinaciones sin repetición")
    print("2. Ejemplo con letras y números")
    print("3. Combinaciones con repetición")
    print("4. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        combinaciones_sin_repeticion()
    elif opcion == "2":
        ejemplo_letras_numeros()
    elif opcion == "3":
        combinaciones_con_repeticion()
    elif opcion == "4":
        print("¡Gracias por usar el programa!")
        break
    else:
        print("Opción no válida. Intente de nuevo.")
```

```python
import math
from itertools import permutations

def permutaciones_sin_repeticion():
    print("\n--- PERMUTACIONES SIN REPETICIÓN ---")
    n = int(input("Ingrese el número total de elementos (n): "))
    r = int(input("Ingrese el número de elementos por grupo (r): "))

    # Fórmula: P(n, r) = n! / (n - r)!
    resultado = math.perm(n, r)

    print(f"El número de permutaciones P({n}, {r}) es: {resultado}")

def ejemplo_permutaciones():
    print("\n--- EJEMPLO DE PERMUTACIONES ---")
    elementos = input("Ingrese los elementos (por ejemplo: ABC): ")
    r = int(input("Ingrese el número de elementos por grupo (r): "))

    lista = list(permutations(elementos, r))

    print(f"\nSe pueden formar {len(lista)} permutaciones:")
    for p in lista:
        print("".join(p))

# Programa principal
while True:
    print("\n=== MENÚ PRINCIPAL ===")
    print("1. Calcular permutaciones numéricas")
    print("2. Ejemplo de permutaciones con caracteres")
    print("3. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        permutaciones_sin_repeticion()
    elif opcion == "2":
        ejemplo_permutaciones()
    elif opcion == "3":
        print("¡Gracias por usar el programa!")
        break
    else:
        print("Opción no válida. Intente de nuevo.")
```

```python
import math

# El usuario define el tipo de problema
print("\nSeleccione el tipo de problema:")
print("1. Combinaciones")
print("2. Permutaciones")

opcion = input("Ingrese su opción: ")

if opcion == "1":
    n = int(input("Ingrese el número total de elementos (n): "))
    r = int(input("Ingrese el número de elementos a elegir (r): "))
    resultado = math.comb(n, r)
    print(f"\nEl número de combinaciones posibles es: {resultado}")
elif opcion == "2":
    n = int(input("Ingrese el número total de elementos (n): "))
    r = int(input("Ingrese el número de elementos a ordenar (r): "))
    resultado = math.perm(n, r)
    print(f"\nEl número de permutaciones posibles es: {resultado}")
else:
    print("Opción no válida. Fin del programa.")
```
